```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
names = []
with open("names.txt",'r') as f:
    names = ['..'+name.strip()+'.' for name in f.readlines()]
len(names)
```

```python
alphabets = ['.'] + [chr(i) for i in range(ord('a'),ord('a')+26)]
i = 0
stoi = {}
for c1 in alphabets:
    for c2 in alphabets:
        stoi[c1+c2] = i
        i += 1

itos = {v:k for k,v in stoi.items()}
```

```python
sord = {char:idx for idx,char in enumerate(alphabets)}
rord = {idx:char for idx,char in enumerate(alphabets)}
```

```python
num = len(stoi)
```

```python
freq = torch.ones((num,27))

for name in names:
    for c1,c2,c3 in zip(name,name[1:],name[2:]):
        pre = c1+c2
        i = stoi[pre]
        j = sord[c3]
        freq[i][j] += 1
```

```python
proba = freq/freq.sum(dim=1,keepdim=True)
```

```python
def sample():
    prev = 0
    word = "."
    while True:
        curr = torch.multinomial(proba[prev],1,True).item()
        if curr == 0:
            return word[1:]
        else:
            word += rord[curr]
            prev = stoi[word[-2:]]
```

```python
sample()
```

```python
log_likelihood = 0.0
n = 0

for name in names:
    for c1,c2,c3 in zip(name,name[1:],name[2:]):
        i,j = stoi[c1+c2],sord[c3]
        log_likelihood += torch.log(proba[i][j])
        n += 1
        
print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
print(f'{nll/n}')
```
